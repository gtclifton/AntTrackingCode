```python
import os
import subprocess
import glob
```

```python
vfolder = '/media/gravishlab/Seagate Expansion Drive/AntTrack/AntTunnel_Longscale/2mm_clear/'
# vfolder = '/media/gravishlab/AntData/AntArena/Rugosity_1_83/'
# vfolder = '/media/gravishlab/AntData/AntBridge/4mm_clear/'
vids = glob.glob(vfolder +'*.avi')
vids = sorted(vids)

print('Num. videos: ', len(vids))
```

```python
nV = 0;
for vid in vids:
    
    outputfile=vid[:len(vid)-9]+'.mp4'
    if not os.path.exists(outputfile):
        commandstring='ffmpeg -i "%s" -vcodec libx264 -pix_fmt yuv420p "%s"' % (vid, outputfile)
    #     commandstring='ffmpeg -i "%s" -vcodec libx264 -pix_fmt gray8 "%s"' % (vid, outputfile) # looks weird
    #     print(commandstring)

        subprocess.call(commandstring, shell=True)
        os.remove(vid)
        print(nV, ' -- ', outputfile)
        nV = nV + 1
    
print('All done!!!')
```

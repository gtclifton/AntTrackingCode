```python
def load_and_analyze_videos_to_df(file_list, buffer):
    vid_locations = ('/').join(file_list[0].split('/')[0:-3])
    col_OI = file_list[0].split('/')[-3][-5:]
    
    trial_info = create_trial_info(file_list)
    tracks = load_in_tracks(file_list)
    tracks = lowpass_filter_tracks(tracks)
    tracks = calculate_v_remove_slow_points(tracks, buffer)
    df = create_dataframe(tracks, trial_info, file_list)
    
    df = remove_redundant_variables(df)
    
    print('\n*** SAVING DF AS PICKLE ***')
    df.to_pickle(vid_locations + '/AllTracks_%s'%col_OI, protocol = 4)#, compression='infer', protocol=4) 
    print('    Done saving pickle')
    
    del df, trial_info, tracks
#     return trial_info


```

```python
def create_trial_info(file_list):
    print('\n*** BUILDING TRIAL LIST ***')
    trial_info = []
    for file in file_list:
        tmps = file.split('/')
        trial_info.append({'substrate': tmps[-2],
                         'date': tmps[-1].split('_')[0],
                         'time': tmps[-1].split('_')[1],
                         'cam': tmps[-1].split('-')[0][-2:],
                         'colony': tmps[-3]})
    del tmps
    print('    Done building trial list')
    return trial_info

```

## Apply lowpass filter to x and y data, find new velocity

butterworth filter
b, a = signal.butter(order, Wn, btype = "low", analog = True)

Wn = cutoff freq / (sample freq/2)


```python
def lowpass_filter_tracks(tracks):
    fps = 239.16
    print('\n*** LOWPASS FILTERING TRACKS ***')
    
    # make butterworth lowpass filter
    b, a = signal.butter(2,0.2,btype='low')

    for cnt, track in enumerate(tracks):
        if track != []:
            for k, obj in track.items():
                if len(obj['x'])>9:
                    obj['xfilt'] = signal.filtfilt(b, a, obj['x'])
                    obj['yfilt'] = signal.filtfilt(b, a, obj['y'])
                    obj['vxfilt'] = np.diff(obj['xfilt'])*fps
                    obj['vyfilt'] = np.diff(obj['yfilt'])*fps
                    obj['vfilt'] = np.sqrt(np.power(obj['vxfilt'],2) + np.power(obj['vyfilt'],2))
    print('    Done lowpass filtering tracks')
    del a, b, fps, cnt, track
    return tracks
```

```python
def calculate_v_remove_slow_points(tracks, buffer):
    print('\n*** CALCULATING VELOCITY ***')
    for cnt, track in enumerate(tracks): # run through each video
        if cnt % 100 == 0:
            print('%i... '%cnt, end = '')
        if track != []:
            for k, obj in track.items(): # run throutgh each track
                trackOI = []
                trackOI = obj.copy()
                if len(trackOI['x'])<10:
                    continue

                # find which frames are close to edge
                wid = 1000
                hei = 550
                edgeidcs = ((trackOI['x'] < buffer) | (trackOI['x'] > wid-(buffer)) | 
                            (trackOI['y'] < buffer) | (trackOI['y'] > hei-buffer))
                ROIidcs = ~edgeidcs

                # find moving points based on distance traveled
                ROIpts = trackOI['vfilt'][ROIidcs[0:-1]]
                ROIfrs = trackOI['frames'][0:-1][ROIidcs[0:-1]]
                Mov_fr = []
                Mov_val = []
                Mov_x = []
                Mov_y = []
                Mov_vx = []
                Mov_vy = []
                Mov_v = []
                # variables
                b_wid = 10
                b_sep = 40
                dist_thres = 30 # pix/window - ~90 pix is 2 mm

                # find average speed over 24 frames (~1 stride)
                wMov_val = []
                wMov_fr = []
                wMov_val_keep = []
                wMov_fr_keep = []
                # variables
                win_wid = 12 # actually half the width
                nfr_cutoff = 20 # of 2*win_wid frames, must have at least this many points to calculate measure

                for fr in ROIfrs:
                    fr = int(fr)

                    if fr - b_sep < 0:
                        continue

                    # what index of all tracked frames does this ROI pt correspond to?
                    idx = np.isin(trackOI['frames'], fr)
    #                 trackOI['vxfilt']=np.append(trackOI['vxfilt'],trackOI['vxfilt'][-1])

                    # what frames do we want to measure position in
                    box1frs = np.array(range(int(fr) - b_wid - b_sep, int(fr) - b_sep))
                    box2frs = np.array(range(int(fr) + b_sep, int(fr) + b_sep + b_wid ))

                    # if box frames aren't in track, do not keep point
                    box1idcs = np.isin(trackOI['frames'], box1frs) #np.searchsorted(trackOI['frames'], box1frs)
                    box2idcs = np.isin(trackOI['frames'], box2frs) 
                    if ~np.any(box1idcs) | ~np.any(box2idcs):
                        # print('one of box frames is not in track - removing fr ', fr)
                        continue

                    # find ave location of ant in time box
                    box1x = np.mean(trackOI['xfilt'][box1idcs])
                    box1y = np.mean(trackOI['yfilt'][box1idcs])
                    box2x = np.mean(trackOI['xfilt'][box2idcs])
                    box2y = np.mean(trackOI['yfilt'][box2idcs])
                    dist_traveled = np.sqrt(np.power(box1x-box2x,2) + np.power(box1y-box2y,2))

                    # append info if is moving point
                    Mov_fr = np.append(Mov_fr, fr)
                    Mov_val = np.append(Mov_val, dist_traveled)
                    Mov_x = np.append(Mov_x, trackOI['xfilt'][idx])
                    Mov_y = np.append(Mov_y, trackOI['yfilt'][idx])
                    if np.where(idx) != len(trackOI['frames']):
                        Mov_vx = np.append(Mov_vx, trackOI['vxfilt'][idx[0:-1]])
                        Mov_vy = np.append(Mov_vy, trackOI['vyfilt'][idx[0:-1]])
                        Mov_v = np.append(Mov_v, trackOI['vfilt'][idx[0:-1]])

                        
                    # find average speed in window
                    winfrs = np.array(range(fr - win_wid, fr + win_wid))
                    winidcs = np.isin(trackOI['frames'], winfrs)

                    # if enough frames in window, calculate average
                    if np.sum(winidcs) < nfr_cutoff:
                        continue   
                    win_medspeed = np.median(trackOI['vfilt'][winidcs[0:-1]])
                    win_meanspeed = np.mean(trackOI['vfilt'][winidcs[0:-1]])
                    wMov_fr = np.append(wMov_fr, fr)
                #     wMov_val = np.append(wMov_val, win_medspeed)
                    wMov_val = np.append(wMov_val, win_meanspeed)

                if np.array(Mov_fr).any(): # remove any stationary behaviors using dist traveled cutoff above
                    wMov_val_keep = wMov_val[[wfr in Mov_fr for wfr in wMov_fr]] # vals when measure is above threshold
                    wMov_fr_keep = wMov_fr[[wfr in Mov_fr for wfr in wMov_fr]] # frs when measure is above threshold

                obj['frames_final'] = Mov_fr
                obj['x_final'] = Mov_x # filtered velocity but only for moving points
                obj['y_final'] = Mov_y
                obj['vx_final'] = Mov_vx
                obj['vy_final'] = Mov_vy
                obj['v_final'] = Mov_v
                obj['dist_90fr'] = Mov_val
                obj['movave_v'] = wMov_val_keep
                obj['movave_fr'] = wMov_fr_keep

    del box1frs, box1idcs, box2frs, box2idcs, box1x, box1y, box2x, box2y, winfrs, winidcs, win_medspeed, win_meanspeed, wMov_fr, wMov_val, wMov_val_keep, wMov_fr_keep
    del Mov_fr, Mov_val, Mov_x, Mov_y, Mov_vx, Mov_vy, Mov_v, dist_traveled, ROIfrs, ROIidcs, ROIpts, win_wid, nfr_cutoff, edgeidcs
    del b_sep, b_wid, dist_thres
    print('\n    Done calculating whole ant velocities')
    return tracks
```

```python
def create_dataframe(tracks, trial_info, file_list):
    all_flights = []
    print('\n*** CREATING DATAFRAME ***')
    for track, trial, video in zip(tracks, trial_info, file_list):
        if track != []:
            for k, obj in track.items():

                # if no x_final etc. hasn't been analyzed bc was too short to filter
                if not 'x_final' in obj.keys():
                    continue

                # if x_final empty no moving points were found in trial
                if len(obj['x_final']) < 1:
                    continue
                    
                tmp = copy.deepcopy(obj)
                tmp['ID'] = k
                tmp['video'] = video
                tmp['substrate'] = trial['substrate']
                tmp['colony'] = trial['colony']
                tmp['time'] = trial['time'] 
                tmp['date'] = trial['date']
                tmp['datetime'] = datetime.datetime.strptime(trial['date']+trial['time'], "%Y%m%d%H%M%S")


                # FOR NOW THESE ARE JUST PIX/S BUT COULD MAKE THEM MM/S ETC.
                tmp['x_kal'] = tmp['x'].tolist() #*pix2cm
                tmp['y_kal'] = tmp['y'].tolist() #*pix2cm
                tmp['x_raw'] = tmp['xfilt'].tolist() #*pix2cm
                tmp['y_raw'] = tmp['yfilt'].tolist() #*pix2cm
                tmp['x'] = tmp['x_final'].tolist() #*pix2cm
                tmp['y'] = tmp['y_final'].tolist() #*pix2cm
                tmp['vx'] = tmp['vx_final'].tolist() #*v_multiplier
                tmp['vy'] = tmp['vy_final'].tolist() #*v_multiplier
                tmp['v'] = tmp['v_final'].tolist() #*v_multiplier
                tmp['v_movave'] = tmp['movave_v'].tolist() #*v_multiplier

    #             tmp['median_v'] = np.median(tmp['v'][inds])
    #             tmp['median_vx'] = np.median(tmp['vx'][inds])
    #             tmp['median_vy'] = np.median(tmp['vy'][inds])
    #             tmp['median_area'] = np.median(tmp['area'][inds])

                tmp['median_v'] = np.median(tmp['v_final'])
                tmp['median_vx'] = np.median(tmp['vx_final'])
                tmp['median_vy'] = np.median(tmp['vy_final'])
    #             tmp['median_area'] = np.median(tmp['area'][inds])

                tmp['sinuosity'] = np.sum(np.sqrt(np.diff(tmp['x_final'])**2 + np.diff(tmp['y_final'])**2)) /   \
                                        np.sqrt((tmp['x_final'][0] - tmp['x_final'][-1])**2 + (tmp['y'][0] - tmp['y_final'][-1])**2)


                all_flights.append(tmp)

    df = pd.DataFrame(all_flights)
    nfrs = np.sum([len(tr['frames']) for tr in all_flights])
    print('Number of tracks: ', len(all_flights), ' - Number of frames: ', nfrs)
    print('    Done creating dataframe')
    del all_flights
    del tracks, tmp, obj
    return df
```

## Remove redundant variables in dataframes

```python
def remove_redundant_variables(df):
    columns_to_drop = [ 'xfilt','yfilt', 'x_final', 'y_final', 'vx_final', 'vy_final', 'v_final']
    # xfilt/yfilt = x_raw/y_raw
    # x_final/y_final = x/y
    # vx_final/vy_final/v_final = vx/vy/v
    for colmn in columns_to_drop:
        if colmn in df: # remove columns if already exist
            df = df.drop(colmn, axis = 1)
    return df
```

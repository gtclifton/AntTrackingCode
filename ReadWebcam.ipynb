```python
import numpy as np
import cv2
import time
import os
import glob
import datetime

# print opencv version
ver = cv2.__version__
# print('---OpenCV Version: ' + ver)

```

```python

# Define webcam item
cap = cv2.VideoCapture(0)
# print('Initial brightness: %.2f' % (cap.get(11)))
# print('Initial contrast: %.2f' % (cap.get(12)))
# print('Initial gain: %.2f' % (cap.get(15)))
# cap.set(cv2.CAP_PROP_AUTOFOCUS, 100) # turn the autofocus off
# focus = 100  # min: 0, max: 255, increment:5
# cap.set(28, focus) 
# cap.set(4,960)
# cap.set(5,600)
cap.set(11,0.8) # brightness
cap.set(12,.2) # contrast
# cap.set(15, ) # gain
# cap.set(16, .5) # exposure
FrNum = 0 

# Test with pre-recorded video
# cap = cv2.VideoCapture('Test.avi')

# Set-up background subtraction
fgbg = cv2.createBackgroundSubtractorMOG2(history = 50,)
# help(cv2.createBackgroundSubtractorMOG2)

# Set-up blob detection
params = cv2.SimpleBlobDetector_Params()
params.minThreshold = 0;
params.maxThreshold = 256;
params.filterByArea = True
params.filterByCircularity = False
params.minArea = 30
params.maxArea = 2000
detector = cv2.SimpleBlobDetector_create(params)
# help(cv2.SimpleBlobDetector)

while(True):

    # Capture frame-by-frame
    ret, frame = cap.read()
    if not ret:
        break
#     print(FrNum)
    FrNum = FrNum + 1
        
    cv2.imshow('raw',frame)
    # crop frame to ROI
    roi = frame[140:330, 50:550]
#     cv2.imshow('crop',roi)
    
    # Apply background subtraction
    fgmask = fgbg.apply(roi)
    
    # Display normal frame
    cv2.imshow('raw', frame)
    cv2.imshow('ROI', roi)
    
    # Display the background frame
#     cv2.imshow('BackgroundSubtraction',fgmask)
    
    if FrNum > 100:
#         print('*****Triggering System online: ', time.ctime())
    
         # Detect bsed on number of pixels diff from background
        ret,thresholded = cv2.threshold(fgmask, 150, 255, cv2.THRESH_BINARY)
        kernel = cv2.getStructuringElement( cv2.MORPH_RECT, (3,3))
        opening = cv2.morphologyEx(thresholded, cv2.MORPH_OPEN, kernel)
        cv2.imshow('Opened image',opening)
    
        total=opening.sum()
        if total>100:
#             print(total)
        
            # make trigger txt file
            t_file = open("/media/gravishlab/Seagate Expansion Drive/AntTrack/Trigger.txt","w")
            t_file.close()

            # save image of webcam to see if lights are on
            currtime = time.strftime("%Y,%m,%d,%H,%M,%S").split(',')
            sname = currtime[0] + currtime[1] + currtime[2] +'_' + currtime[3] + currtime[4] + currtime[5]
            cv2.imwrite("/media/gravishlab/Seagate Expansion Drive/AntTrack/%s.jpg" % sname, frame)

            # tell the world you've triggered the cams
            print('---Recording Triggered: ', time.ctime())
            
            # find previous jpg and what corresponding avi name should be
#             alljpg = sorted(glob.glob("/media/gravishlab/Seagate Expansion Drive/AntTrack/*.jpg"))
#             lastjpg = alljpg[-2]
#             jpgtime = lastjpg.split('/')[-1][9:15]
#             vtime1 = (datetime.datetime.strptime(jpgtime,'%H%M%S') + datetime.timedelta(seconds=3)).strftime('%H%M%S')
#             vtime2 = (datetime.datetime.strptime(jpgtime,'%H%M%S') + datetime.timedelta(seconds=4)).strftime('%H%M%S')
#             vname1 = lastjpg[:-10] + vtime1 + '_16276718-0000.avi'
#             vname2 = lastjpg[:-10] + vtime2 + '_16276718-0000.avi'
                
#             # if no avi that corresponds to last jpg?
#             if (not os.path.exists(vname1)) and (not os.path.exists(vname2)):
        
#                 # delete previous jpg
#                 os.remove(lastjpg)
            
            
            time.sleep(40)
            
            

            # RecordVideos exe will remove Trigger file
#           os.remove("/home/gravishlab/Documents/AntTrack/Trigger.txt")
            FrNum = 0;
            
        
        # Detect blobs
#         keypoints = detector.detect(255-fgmask)
#         for keypoint in keypoints:    
#     #         print(keypoint.size)
#             if keypoint.size > 5:
#                 #t_file = open("/home/gravishlab/Documents/AntTrack/Trigger.txt","w")
#                 t_file = open("/media/gravishlab/Seagate Expansion Drive/AntTrack/Trigger.txt","w")
#                 t_file.close()
#                 print('---Recording Triggered: ', time.ctime())
#                 time.sleep(20)
#                 # RecordVideos exe will remove Trigger file
# #                 os.remove("/home/gravishlab/Documents/AntTrack/Trigger.txt")
#                 FrNum = 0;
    
#         # display videos with blobs outlined in red
#         mask_with_keypoints = cv2.drawKeypoints(fgmask, keypoints, np.array([]), (0,0,255), cv2.DRAW_MATCHES_FLAGS_DRAW_RICH_KEYPOINTS)
#         cv2.imshow('mask_keypoints',mask_with_keypoints)
#         im_with_keypoints = cv2.drawKeypoints(roi, keypoints, np.array([]), (0,0,255), cv2.DRAW_MATCHES_FLAGS_DRAW_RICH_KEYPOINTS)
#         cv2.imshow('keypoints',im_with_keypoints)
    
    
    # how close program
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break
        
# When everything done, release the capture
cap.release()
cv2.destroyAllWindows()
    
```

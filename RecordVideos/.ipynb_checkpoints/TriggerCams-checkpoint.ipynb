```python
# flag of if program should run
flag = 1

# run RecordVideos.cpp code
def func1():
#     global rocket
    print('---Starting RecordVideos')
    os.system('~/spinnaker_1_0_0_262/bin/RecordVideos')
    time.sleep(8)

# run Webcam
def func2():
    print('---Starting webcam')
    %run ReadWebcam.ipynb
    time.sleep(8)
    
# track all newly recorded files
def func3():
    print('---Tracking recorded videos')
    # move any fully tracked and associated files
    %run MoveToFolder.ipynb
    
    # associate any tracked files then move
    %run _2_Run_Data_Associator.ipynb
    %run MoveToFolder.ipynb
    
#     # track, associate, then move any files
#     %run DeleteNoAntVids.ipynb
# #    %run _1_Track_all_videos.ipynb
# #    %run _2_Run_Data_Associator.ipynb
#     %run MoveToFolder.ipynb

    
# thread of normal function
def mainpgm():
    print('Start of program!')
    print('')
    print('Starting...')
    p1 = Process(target=func1)
    p1.start()
    p2 = Process(target=func2)
    p2.start()
#     p3 = Process(target=func3)
#     p3.start()
    time.sleep(4)
    
    # start timer
    start_time=time.time()
    
    while(flag):
        elapsed_time=time.time()-start_time
        
        if(elapsed_time>1200): # restart every __ seconds
            print('')
            print('Exiting...')
            
            # find process index of webcam and kill
            proc = subprocess.Popen(["fuser /dev/video0"], stdout=subprocess.PIPE, shell=True)
            (out, err) = proc.communicate()
            killcommand = 'kill -9' + out.decode('UTF-8')
            os.system(killcommand)
            print('---closed webcam, PID:', out.decode('UTF-8'))
            
            # give RecordVideos time to finish
            time.sleep(40)
            
            # close RecordVideos
            %run KillRecordVideos.ipynb
            
            print('')
            print('Restarting...')
            p1 = Process(target=func1)
            p1.start()
            p2 = Process(target=func2)
            p2.start()
#             p3 = Process(target=func3)
#             p3.start()
            time.sleep(4)
            
            # start timer
            start_time=time.time()


            
            
# thread of user input
def get_input():
    global flag
    print('')
    keystrk=input('Press a key to exit: ')
    # thread doesn't continue until key is pressed
    flag=False
    print('You pressed: ', keystrk, ' -- flag is now:', flag)
    print('')

    print('Exiting...')
    
    # find process index of webcam and kill
    proc = subprocess.Popen(["fuser /dev/video0"], stdout=subprocess.PIPE, shell=True)
    (out, err) = proc.communicate()
    killcommand = 'kill -9' + out.decode('UTF-8')
    os.system(killcommand)
    print('---closed webcam, PID:', out.decode('UTF-8'))
            
    # give RecordVideos time to finish
    time.sleep(30)
            
    # close RecordVideos
    %run KillRecordVideos.ipynb   
    
    
    
        
if __name__=='__main__':
    n=threading.Thread(target=mainpgm)
    i=threading.Thread(target=get_input)
    n.start()
    i.start()
    

    
#     if flag == "0":
#         print('Exiting...')
#         %run KillRecordVideos.ipynb
        
#         # find process index of webcam and kill
#         proc = subprocess.Popen(["fuser /dev/video0"], stdout=subprocess.PIPE, shell=True)
#         (out, err) = proc.communicate()
#         killcommand = 'kill -9' + out.decode('UTF-8')
#         os.system(killcommand)
#         print('---closed webcam, PID:', out.decode('UTF-8'))

```
